```python

emotion_suggestions = {
    "happy": "Keep up the great vibes! Maybe share the joy with your team 😊",
    "sad": "Take a deep breath. You're stronger than you think. Want to try a small relaxing task?",
    "angry": "Try a quick walk or a few deep breaths. Let’s shift the energy to something positive 💪",
    "surprise": "Something unexpected? Take a moment to absorb it — you're doing well.",
    "neutral": "All set! Let’s pick a steady and productive task today.",
    "fear": "It’s okay to feel unsure. You got this — try something familiar to gain confidence.",
    "disgust": "Something bothering you? Let’s clean the slate and focus on what inspires you.",
    "calm": "You're in a great headspace. Let’s keep the momentum going!"
}
```

```python
from transformers import pipeline
from fer import FER
import gradio as gr
import numpy as np
import librosa

text_emotion = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=False)
face_detector = FER(mtcnn=True)


emotion_suggestions = {
    "joy": "You seem happy! Keep up the great work and spread positivity 😊",
    "happy": "Keep up the great vibes! Maybe share the joy with your team 😊",
    "sad": "Take a deep breath. You're stronger than you think. Want to try a small relaxing task?",
    "anger": "Try a quick walk or a few deep breaths. Let’s shift the energy to something positive 💪",
    "angry": "Take a deep breath. Consider doing something calming for a bit.",
    "surprise": "Something unexpected? Take a moment to absorb it — you're doing well.",
    "neutral": "All set! Let’s pick a steady and productive task today.",
    "fear": "It’s okay to feel unsure. You got this — try something familiar to gain confidence.",
    "disgust": "Something bothering you? Let’s clean the slate and focus on what inspires you.",
    "calm": "You're in a great headspace. Let’s keep the momentum going!",
    "love": "That's sweet! Let your positive energy inspire others too.",
    "confusion": "Let’s simplify things. Maybe try breaking the task down.",
    "boredom": "Maybe try a creative or new task to shake things up!"
}


def analyze_text_emotion(text):
    result = text_emotion(text)[0]
    label = result['label'].strip().lower()
    suggestion = emotion_suggestions.get(label, f"No specific suggestion for '{label}', but you're doing great!")
    return f"📝 Text Emotion: {label.capitalize()} ({round(result['score']*100, 2)}%)\n💡 Suggestion: {suggestion}"


def analyze_frame(frame):
    result = face_detector.top_emotion(frame)
    if result:
        emotion, score = result
        label = emotion.strip().lower()
        suggestion = emotion_suggestions.get(label, f"No specific suggestion for '{label}', but you're doing great!")
        return f" Facial Emotion: {label.capitalize()} ({round(score*100, 2)}%)\n💡 Suggestion: {suggestion}"
    return "No face detected. Please try again."


def analyze_audio(audio_file):
    try:
        y, sr = librosa.load(audio_file)
        pitch = librosa.pitch_tuning(librosa.yin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'), sr=sr))
        emotion = "calm" if pitch < 0 else "angry" if pitch > 0.5 else "neutral"
    except Exception as e:
        print(f"Error processing audio: {e}")
        return "Error analyzing audio."
    suggestion = emotion_suggestions.get(emotion, f"No specific suggestion for '{emotion}', but you're doing great!")
    return f"🎙️ Audio Emotion: {emotion.capitalize()}\n💡 Suggestion: {suggestion}"


def confirm_capture(image):
    if image is not None:
        return "✅ Photo captured successfully!"
    return "⚠️ Please capture or upload a photo."


text_interface = gr.Interface(
    fn=analyze_text_emotion,
    inputs=gr.Textbox(label="Enter Your Feelings (Text)"),
    outputs=gr.Text(label="Result"),
    title="Text-Based Emotion Detection"
)


with gr.Blocks() as face_interface:
    gr.Markdown("## Facial Emotion Detection")

    webcam_input = gr.Image(label="Upload or Capture Face Image", type="numpy")

    capture_btn = gr.Button("📸 Capture Snapshot")
    capture_status = gr.Textbox(label="Capture Status")

    capture_btn.click(fn=confirm_capture, inputs=webcam_input, outputs=capture_status)

    analyze_btn = gr.Button("Analyze Emotion")
    result_output = gr.Text(label="Result")

    analyze_btn.click(fn=analyze_frame, inputs=webcam_input, outputs=result_output)


voice_interface = gr.Interface(
    fn=analyze_audio,
    inputs=gr.Audio(label="Upload or Record Your Voice", type="filepath"),
    outputs=gr.Text(label="Result"),
    title="Voice Emotion Detection"
)


demo = gr.TabbedInterface(
    [text_interface, face_interface, voice_interface],
    tab_names=["Text", "Face", "Voice"]
)


demo.launch()
```
